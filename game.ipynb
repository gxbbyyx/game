```python
import tkinter as tk
import random
from PIL import Image, ImageTk

window = tk.Tk()
window.title("RNG Game")

window_width = 900
window_height = 900
screen_width = window.winfo_screenwidth()
screen_height = window.winfo_screenheight()
x = (screen_width - window_width) // 2
y = (screen_height - window_height) // 2
window.geometry(f"{window_width}x{window_height}+{x}+{y}")

characters = [
    {"name": "Fleurdelys", "odds": 500_000, "image": "images/Fleurdelys.jpg"},
    {"name": "Carlotta", "odds": 676_000, "image": "images/carlotta.jpg"},
    {"name": "Cartethyia", "odds": 666_000, "image": "images/cartethyia.jpg"},
    {"name": "Hatsune Miku", "odds": 400_000, "image": "images/hatsunemiku.jpg"},
    {"name": "Ada Wong", "odds": 145_000, "image": "images/adawong.jpg"},
    {"name": "Elysia", "odds": 200_000, "image": "images/elysia.jpg"},
    {"name": "Aemeath", "odds": 55_000, "image": "images/aemeath.jpg"},
    {"name": "The Herta", "odds": 500_000, "image": "images/herta.jpg"},
    {"name": "The Dahlia", "odds": 250_000, "image": "images/dahlia.jpg"},
    {"name": "Kiana Kaslana", "odds": 150_000, "image": "images/kiana.jpg"},
    {"name": "Raiden Mei", "odds": 130_000, "image": "images/raidenmei.jpg"},
    {"name": "Raiden Shogun", "odds": 195_000, "image": "images/raidenshogun.jpg"},
    {"name": "Alice", "odds": 156_000, "image": "images/alice.jpg"},
    {"name": "Phrolova", "odds": 99_000, "image": "images/phrolova.jpg"},
    {"name": "Wedding Yae Miko & Raiden Shogun", "odds": 97_500, "image": "images/wedyaeraiden.jpg"},
    {"name": "Augusta", "odds": 95_000, "image": "images/augusta.jpg"},
    {"name": "Iuno", "odds": 80_000, "image": "images/iuno.jpg"},
    {"name": "Furina", "odds": 40_000, "image": "images/furina.jpg"},
    {"name": "Yelan", "odds": 30_000, "image": "images/yelan.jpg"},
    {"name": "Clorinde", "odds": 24_000, "image": "images/clorinde.jpg"},
    {"name": "Lauma", "odds": 15_000, "image": "images/lauma.jpg"},
    {"name": "Black Cat", "odds": 12_000, "image": "images/blackcat.jpg"},
    {"name": "Keqing", "odds": 4_500, "image": "images/keqing.jpg"},
    {"name": "Flora", "odds": 1_000, "image": "images/flora.jpg"},
    {"name": "Bloom", "odds": 1_000, "image": "images/bloom.png"},
    {"name": "Shorekeeper", "odds": 150, "image": "images/shorekeeper.jpg"},
    {"name": "Cantarella", "odds": 80, "image": "images/cantarella.jpg"},
    {"name": "Lynae", "odds": 55, "image": "images/lynae.jpg"},
    {"name": "Musa", "odds": 50, "image": "images/musa.jpg"},
    {"name": "Zibai", "odds": 20, "image": "images/zibai.jpg"},
    {"name": "Rogue", "odds": 10, "image": "images/rogue.jpg"},
    {"name": "Hysilens", "odds": 5, "image": "images/hysilens.jpg"},
    {"name": "Cinnamoroll", "odds": 1, "image": "images/cinnamoroll2.jpg"}
]

roll_history = []
kept_cards = []
roll_count = 0
auto_rolling = False
luck = 1
current_roll = None

def load_image(image_path):
    try:
        img = Image.open(image_path)
        img = img.resize((320, 450))
        return ImageTk.PhotoImage(img)
    except:
        return None

def update_roll_history():
    roll_history_listbox.delete(0, tk.END)
    for item in roll_history:
        roll_history_listbox.insert(tk.END, f"{item['name']} (1 in {item['odds']:,})")

def update_kept_cards():
    kept_listbox.delete(0, tk.END)
    for item in kept_cards:
        kept_listbox.insert(tk.END, f"{item['name']} (1 in {item['odds']:,})")

def roll_character():
    weights = [luck / character["odds"] for character in characters]
    return random.choices(characters, weights=weights, k=1)[0]

def do_roll():
    global roll_count, current_roll

    rolled = roll_character()
    current_roll = rolled
    roll_count += 1
    roll_history.append(rolled)

    result_label.config(
        text=f"You rolled: {rolled['name']} (1 in {rolled['odds']:,})"
    )
    counter_label.config(text=f"Rolls: {roll_count}")

    img = load_image(rolled["image"])
    if img:
        image_label.config(image=img, text="")
        image_label.image = img
    else:
        image_label.config(image="", text="Image not found")
        image_label.image = None

    update_roll_history()
    update_keep_button()

def update_keep_button():
    if current_roll:
        keep_button.config(state="normal")
    else:
        keep_button.config(state="disabled")

def keep_current_card():
    if current_roll is None:
        return

    kept_cards.append(current_roll)
    update_kept_cards()

def auto_roll_loop():
    if auto_rolling:
        do_roll()
        window.after(500, auto_roll_loop)

def start_auto_roll():
    global auto_rolling
    if not auto_rolling:
        auto_rolling = True
        auto_roll_loop()

def stop_auto_roll():
    global auto_rolling
    auto_rolling = False

def increase_luck():
    global luck
    luck += 1
    luck_label.config(text=f"Luck: {luck}x")

def reset_luck():
    global luck
    luck = 1
    luck_label.config(text="Luck: 1x")

def view_kept_card(event):
    selection = kept_listbox.curselection()
    if not selection:
        return

    index = selection[0]
    card = kept_cards[index]

    result_label.config(
        text=f"Viewing kept card: {card['name']} (1 in {card['odds']:,})"
    )

    img = load_image(card["image"])
    if img:
        image_label.config(image=img, text="")
        image_label.image = img
    else:
        image_label.config(image="", text="Image not found")
        image_label.image = None

# ----- LEFT SIDE -----
left_frame = tk.Frame(window)
left_frame.pack(side="left", fill="both", expand=True, padx=15, pady=15)

title_label = tk.Label(left_frame, text="RNG", font=("Arial", 22, "bold"))
title_label.pack(pady=10)

image_label = tk.Label(left_frame, text="Roll to see a character", relief="solid", bd=2)
image_label.pack(pady=10)

result_label = tk.Label(left_frame, text="Press ROLL to begin", font=("Arial", 14))
result_label.pack(pady=8)

counter_label = tk.Label(left_frame, text="Rolls: 0", font=("Arial", 12))
counter_label.pack(pady=4)

luck_label = tk.Label(left_frame, text="Luck: 1x", font=("Arial", 12, "bold"))
luck_label.pack(pady=4)

button_frame = tk.Frame(left_frame)
button_frame.pack(pady=10)

roll_button = tk.Button(button_frame, text="ROLL", font=("Arial", 14, "bold"), width=14, command=do_roll)
roll_button.grid(row=0, column=0, padx=5, pady=5)

start_button = tk.Button(button_frame, text="Start Auto Roll", font=("Arial", 12), width=14, command=start_auto_roll)
start_button.grid(row=0, column=1, padx=5, pady=5)

stop_button = tk.Button(button_frame, text="Stop Auto Roll", font=("Arial", 12), width=14, command=stop_auto_roll)
stop_button.grid(row=0, column=2, padx=5, pady=5)

increase_luck_button = tk.Button(button_frame, text="Increase Luck", font=("Arial", 12), width=14, command=increase_luck)
increase_luck_button.grid(row=1, column=0, padx=5, pady=5)

reset_luck_button = tk.Button(button_frame, text="Reset Luck", font=("Arial", 12), width=14, command=reset_luck)
reset_luck_button.grid(row=1, column=1, padx=5, pady=5)

keep_button = tk.Button(button_frame, text="KEEP CARD", font=("Arial", 12, "bold"), width=14, command=keep_current_card, state="disabled")
keep_button.grid(row=1, column=2, padx=5, pady=5)

# ----- RIGHT SIDE -----
right_frame = tk.Frame(window)
right_frame.pack(side="right", fill="both", expand=True, padx=15, pady=15)

history_title = tk.Label(right_frame, text="Roll History", font=("Arial", 16, "bold"))
history_title.pack(pady=5)

history_frame = tk.Frame(right_frame)
history_frame.pack(fill="both", expand=True, pady=5)

history_scrollbar = tk.Scrollbar(history_frame)
history_scrollbar.pack(side="right", fill="y")

roll_history_listbox = tk.Listbox(
    history_frame,
    font=("Arial", 11),
    yscrollcommand=history_scrollbar.set
)
roll_history_listbox.pack(side="left", fill="both", expand=True)
history_scrollbar.config(command=roll_history_listbox.yview)

kept_title = tk.Label(right_frame, text="Kept Cards", font=("Arial", 16, "bold"))
kept_title.pack(pady=5)

kept_frame = tk.Frame(right_frame)
kept_frame.pack(fill="both", expand=True, pady=5)

kept_scrollbar = tk.Scrollbar(kept_frame)
kept_scrollbar.pack(side="right", fill="y")

kept_listbox = tk.Listbox(
    kept_frame,
    font=("Arial", 11),
    yscrollcommand=kept_scrollbar.set
)
kept_listbox.pack(side="left", fill="both", expand=True)
kept_scrollbar.config(command=kept_listbox.yview)

kept_listbox.bind("<<ListboxSelect>>", view_kept_card)

window.mainloop()
```
